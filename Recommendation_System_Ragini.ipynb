#Problem Statement
This project simulates a recommendation system using batch and real-time processing.
It provides personalized recommendations based on user history and real-time updates.

#SAMPLE DATA


```python
# User history (batch data)
user_history = {
    "user1": ["movie1", "movie2"],
    "user2": ["movie2", "movie3"]
}

# Item popularity
item_popularity = {
    "movie1": 5,
    "movie2": 10,
    "movie3": 7,
    "movie4": 15
}
```

#BATCH PROCESSING

```python
def batch_recommend(user):
    watched = user_history.get(user, [])
    recommendations = []

    for item in item_popularity:
        if item not in watched:
            recommendations.append((item, item_popularity[item]))

    recommendations.sort(key=lambda x: x[1], reverse=True)
    return [item for item, _ in recommendations[:3]]

print("Batch Recommendation:", batch_recommend("user1"))
```

#REAL-TIME PROCESSING

```python
def real_time_update(user, new_item):
    if user not in user_history:
        user_history[user] = []

    user_history[user].append(new_item)

    if new_item in item_popularity:
        item_popularity[new_item] += 1
    else:
        item_popularity[new_item] = 1

# simulate real-time
real_time_update("user1", "movie4")

print("After Real-time Update:", batch_recommend("user1"))
```

# QUEUE (Kafka Simulation)

```python
from queue import Queue

event_queue = Queue()

def send_event(user, item):
    event_queue.put((user, item))

def process_events():
    while not event_queue.empty():
        user, item = event_queue.get()
        real_time_update(user, item)

# simulate event
send_event("user2", "movie4")
process_events()

print("Queue Processed:", batch_recommend("user2"))
```

# CACHE (Redis Simulation)

```python
cache = {}

def get_recommendation(user):
    if user in cache:
        print("Cache Hit")
        return cache[user]

    print("Cache Miss")
    result = batch_recommend(user)
    cache[user] = result
    return result

print(get_recommendation("user1"))
print(get_recommendation("user1"))
```
